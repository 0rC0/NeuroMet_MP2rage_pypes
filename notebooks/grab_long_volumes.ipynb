```python
spaths = glob.glob('/media/drive_s/AG/AG-Floeel-Imaging/02-User/NEUROMET_BIDS/NeuroMET/derivatives/NeuroMET_FS/stats_long/tmp/*reconall.long*')
```

```python
main_df = pd.DataFrame()
for i in spaths:
    try:
        sub = os.path.basename(i)
        print(sub)
        df = pd.DataFrame()
        for j in ['mri/rh.hippoSfVolumes-T1.long.v21.txt', 'mri/rh.amygNucVolumes-T1.long.v21.txt', 'lh.hippoSfVolumes-T1.long.v21.txt', 'lh.amygNucVolumes-T1.long.v21.txt']:
            data = os.path.join(i, j)
            if os.path.isfile(data):
                #print(data)
                df_tmp =pd.read_csv(data, sep=' ', header=None)
                prefix = 'rh_' if 'rh.' in j else 'lh_'
                df_tmp[0] =df_tmp[0].apply(lambda x: prefix + x)
                df_tmp.columns = ['ROI',sub]
                df = df.append(df_tmp)
                #print(df)
            else:
                print(data)
        #print(df)
        if len(main_df) == 0:
            main_df  = df.set_index('ROI')
        else:
            main_df = main_df.join(df.set_index('ROI'))
    except:
        
        pass
```
